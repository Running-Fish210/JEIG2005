```python
import numpy as np

def F1(W, y, A, E, I, L, P):
    '''
    已知载荷求应力，挠度
    '''
    u = np.sqrt(P / (E * I))
    s_max = P/A + W/(2*u) * np.tan(u * L / 2) * y / I
    v_max = W/(2*P) * (np.tan(u * L / 2) / u - L/2)
    
    print(f'最大位移 = {v_max*1e3}mm')
    print(f'最大应力 = {s_max/1e6}MPa')

def F2(W, y, A, E, I, L, s_max):
    '''
    已知应力求载荷
    '''
    P1, P2, P3 = Pc, Pc/2, 0
    u = np.sqrt(P2 / (E * I))
    s = P2/A + W/(2*u) * np.tan(u * L / 2) * y / I
    while np.abs(s-s_max) > 0.01:
        if s > s_max:
            P1 = P2
            P2 = (P2 + P3)/2 
            u = np.sqrt(P2 / (E * I))
            s = P2/A + W/(2*u) * np.tan(u * L / 2) * y / I
        elif s < s_max:
            P3 = P2
            P2 = (P1 + P2)/2
            u = np.sqrt(P2 / (E * I))
            s = P2/A + W/(2*u) * np.tan(u * L / 2) * y / I
    v_max = W/(2*P2) * (np.tan(u * L / 2) / u - L/2)    
    print(f'最大载荷为 = {P2/1e3} = {np.round(P2/1e3,3)}kN')
    print(f'最大挠度 = {v_max*1e3} = {np.round(v_max*1e3)}mm')

def F3(W, y, A, E, I, L, v_max):
    '''
    已知挠度求载荷
    '''
    P1, P2, P3 = Pc, Pc/2, 0
    u = np.sqrt(P2 / (E * I))
    v = W/(2*P2) * (np.tan(u * L / 2) / u - L/2)
    while np.abs(v-v_max)*1e3 > 0.01:
        if v > v_max:
            P1 = P2
            P2 = (P2 + P3)/2 
            u = np.sqrt(P2 / (E * I))
            v = W/(2*P2) * (np.tan(u * L / 2) / u - L/2)
        elif v < v_max:
            P3 = P2
            P2 = (P1 + P2)/2
            u = np.sqrt(P2 / (E * I))
            v = W/(2*P2) * (np.tan(u * L / 2) / u - L/2)
    s_max = P2/A + W/(2*u) * np.tan(u * L / 2) * y / I  
    print(f'最大载荷为 = {P2/1e3} = {np.round(P2/1e3,3)}kN')
    print(f'最大应力为 = {s_max*1e3} = {np.round(s_max*1e3,3)}MPa')

'''-------------------------—---分割线----------------------------------'''

pi = np.pi
A = pi*0.025**2/4
I = pi*0.025**4/64
E = 210e9
L = 1.5
W = 40
y = 0.025/2
Pc = pi**2 * E * I / L**2

'''-------------------------—---分割线----------------------------------'''

print(f'欧拉屈曲 Pc = {Pc/1e3} = {np.round(Pc/1e3,3)}kN')

#""" 已知力，求位移和应力 """
#F1(W, y, A, E, I, L, P = 16e3)

#""" 已知最大应力，求最大载荷 """
#F2(W, y, A, E, I, L, s_max = 240e6) 

#""" 已知最大挠度，求最大载荷 """
#F3(W, y, A, E, I, L, v_max = )
```
